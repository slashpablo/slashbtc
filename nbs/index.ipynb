# slashbtc

> Python tooling for inspecting the Bitcoin blockchain.

````python
#| hide
from slashbtc.core import *
````

## Install

```sh
uv add slashbtc
# or
pip install slashbtc
```

## Configure

Put your node's URL and cookie in `.env`:

```
BITCOIN_RPC_URL=http://127.0.0.1:8332
BITCOIN_RPC_COOKIE=.secrets/bitcoin-rpc.cookie
```

`BITCOIN_RPC_COOKIE` may be either a path to Bitcoin Core's `.cookie` file or the raw `user:password` string.

## Use

```python
from slashbtc.core import Chain

chain = Chain()
chain.height()                    # int
chain.block(800_000)              # dict, full tx data
chain.tx("abc...")                # decoded transaction
```

## Develop

This package is built with [nbdev](https://nbdev.fast.ai). Source of truth lives in `nbs/`; the `slashbtc/` Python package is generated.

```sh
uv sync --extra dev          # install nbdev + jupyter into .venv
uv run jupyter lab           # edit notebooks in nbs/
uv run nbdev_export          # regenerate slashbtc/ from notebooks
uv run nbdev_test            # run notebook-based tests
uv run nbdev_readme          # regenerate README.md from index.ipynb
```
